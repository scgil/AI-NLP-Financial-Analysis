# Ratio Analysis

# 1. Recovery of Ratios and their values for a Company from a CSV file

Specify the path and name of the CSV file. Read the file and display the resulting ratios dataframe.

```python
import pandas as pd

pd.set_option('display.max_colwidth', None)
```

```python
path = './Ratios/'
file = 'ratios_'
company_name = 'WaltDisney'
type_file = '.csv'

ratios = pd.read_csv(path+file+company_name+type_file, index_col = 0)

ratios
```

# 2. Perform Ratio Analysis

Perform ratio analysis for each of the ratios and store them in their respective variables.

```python
current_ratio = "The company's Current ratio decreased from 1.08 in 2021 to 1.00 in 2022, indicating a reduction in its liquidity. Both ratios fall within the range of 1.00 to 1.49, suggesting that the company's current assets, including cash, accounts receivable and inventory, are generally sufficient to cover its short-term liabilities. However, it may still be a need for careful management of liquidity to handle unexpected events."
```

```python
quick_ratio = "The company's Quick ratio declined from 1.04 in 2021 to 0.94 in 2022, indicating a reduction in its liquidity. In 2021, the Quick ratio fell within the range of 1.00 to 1.50, indicating that the company had a sufficient buffer of liquid assets (such as cash, marketable securities, and accounts receivable) to cover its short-term liabilities. In 2022, a Quick ratio below 1.00 raises concerns about the company's ability to cover current liabilities without relying on inventory or other less liquid assets."
```

```python
cash_ratio = "The company's Cash ratio decreased from 0.51 in 2021 to 0.40 in 2022, reflecting a reduction in its liquidity. In 2021, the Cash ratio was above 0.40, suggesting an excess of liquidity. This implied that the company had more cash on hand than needed for its short-term liabilities. In 2022, the Cash ratio is in the range between 0.20 and 0.40, indicating a balanced liquidity position. The specified range suggests that the company has a reasonable amount of cash to cover a significant portion of its short-term liabilities and the company is likely to meet its immediate financial obligations without significant challenges."
```

# 3. Add the Ratio Analysis as a column to the dataframe and save the file

Include a new column in the dataframe with the performed ratio analysis. Display the resulting dataframe.

```python
ratios['Analisis'] = [current_ratio, quick_ratio, cash_ratio]

ratios
```

Save the dataframe to a CSV file with the specified path and name.

```python
path = './Analyses/'
file = 'analysis_'
type_file = '.csv'

ratios.to_csv(path+file+company_name+type_file)
```

# 4. Create a dataframe to aggregate all Ratio Analyses into a single file

Create an empty dataframe with the proposed structure for the joint storage of ratio analyses from different companies. Display the structure of the created dataframe.

```python
columns = ['company', 'current_ratio_2021', 'current_ratio_2022', 
           'quick_ratio_2021', 'quick_ratio_2022', 
           'cash_ratio_2021', 'cash_ratio_2022', 'analysis_current_ratio', 'analysis_quick_ratio', 'analysis_cash_ratio']

ratio_analysis = pd.DataFrame(index=[0], columns=columns)

ratio_analysis
```

# 5. Add the current Ratio Analysis to the combined set of Ratio Analyses file

## 5.1 Enter the information from the performed Ratio Analysis into the new structure

Create variables to store the index names, column names, ratio values and ratio analyses from the dataframe containing the actual ratio analysis. Display the resulting dataframe.

```python
index_current_ratio = 'Current_Ratio_' + company_name
index_quick_ratio = 'Quick_Ratio_' + company_name
index_cash_ratio = 'Cash_Ratio_' + company_name

year_X = '2021'
year_Y = '2022'
analysis = 'Analisis'

current_ratio_value_year_X = '{:.2f}'.format(ratios.loc[index_current_ratio, year_X])
current_ratio_value_year_Y = '{:.2f}'.format(ratios.loc[index_current_ratio, year_Y])
quick_ratio_value_year_X = '{:.2f}'.format(ratios.loc[index_quick_ratio, year_X])
quick_ratio_value_year_Y = '{:.2f}'.format(ratios.loc[index_quick_ratio, year_Y])
cash_ratio_value_year_X = '{:.2f}'.format(ratios.loc[index_cash_ratio, year_X])
cash_ratio_value_year_Y = '{:.2f}'.format(ratios.loc[index_cash_ratio, year_Y])
analysis_current_ratio = ratios.loc[index_current_ratio, analysis]
analysis_quick_ratio = ratios.loc[index_quick_ratio, analysis]
analysis_cash_ratio = ratios.loc[index_cash_ratio, analysis]


ratio_analysis['company'] = company_name
ratio_analysis['current_ratio_2021'] = current_ratio_value_year_X
ratio_analysis['current_ratio_2022'] = current_ratio_value_year_Y
ratio_analysis['quick_ratio_2021'] = quick_ratio_value_year_X
ratio_analysis['quick_ratio_2022'] = quick_ratio_value_year_Y
ratio_analysis['cash_ratio_2021'] = cash_ratio_value_year_X
ratio_analysis['cash_ratio_2022'] = cash_ratio_value_year_Y
ratio_analysis['analysis_current_ratio'] = analysis_current_ratio
ratio_analysis['analysis_quick_ratio'] = analysis_quick_ratio
ratio_analysis['analysis_cash_ratio'] = analysis_cash_ratio

ratio_analysis
```

### 5.2.2 Subsequent times: Retrieve the file and add the current analysis

Retrieve the file containing the analysis of all companies and add the current company's ratio analysis. Display the resulting dataframe.

```python
file = 'analyses_ratios'
ratio_analyses = pd.read_csv(path+file+type_file, index_col = 0)

ratio_analyses_all = pd.concat([ratio_analyses, ratio_analysis], ignore_index=True)

ratio_analyses_all.to_csv(path+file+type_file, index=True)

ratio_analyses_all
```
