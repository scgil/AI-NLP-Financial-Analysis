# 2. Cross-Validation samples

## 2.1 Cross-Validation process

Cross-validation process: splitting the data by ratio type and saving the training and test sets in CSV files for subsequent use in the model training and evaluation process.

```python
from sklearn.model_selection import KFold
import os

ratios = sample['ratio'].unique()

num_iterations = 5

save_path = "./Cross_Validation_Sample/Cross_Validation_CSV_Files"

os.makedirs(save_path, exist_ok=True)

for ratio in ratios:
    ratio_df = sample[sample['ratio'] == ratio]
    
    kf = KFold(n_splits=num_iterations, shuffle=True, random_state=42)

    for iteration, (train_index, test_index) in enumerate(kf.split(ratio_df)):
        train_set, test_set = ratio_df.iloc[train_index], ratio_df.iloc[test_index]

        train_file = os.path.join(save_path, f'train_set_{ratio}_{iteration + 1}.csv')
        test_file = os.path.join(save_path, f'test_set_{ratio}_{iteration + 1}.csv')

        train_set.to_csv(train_file, index=False)
        test_set.to_csv(test_file, index=False)

        print(f'Ratio: {ratio}, Iteration: {iteration + 1}, Saving training and test set in {save_path}')
```

# 3. Convert CSV to JSON files with conversation format

## 3.1 CSV to JSON function

Define a function named convert_df_to_json, which takes a DataFrame as input and converts it into a list of JSON-formatted conversations. For each row in the DataFrame, a unique conversation ID is generated using the uuid module. The function creates a dictionary representing a conversation with two messages: one from "human" with the instruction from the DataFrame's "instruction" column, and another from "gpt" with the analysis from the "analysis" column. These conversation dictionaries are then appended to a list called data.

```python
import pandas as pd
import json
import uuid

def convert_df_to_json(df):
    data = []
    for index, row in df.iterrows():
        conversation_id = str(uuid.uuid4())
        conversation = {
            "id": conversation_id,
            "conversations": [
                {"from": "human", "value": row["instruction"]},
                {"from": "gpt", "value": row["analysis"]}
            ]
        }
        data.append(conversation)

    return data
```

## 3.2 Retrieve CSV file

Define the path, file and file type. Read the CSV file into a DataFrame.

```python
path = './Cross_Validation_Sample/Cross_Validation_CSV_Files/'
file = 'test_set_Cash_5'
type_file = '.csv'

dataframe = pd.read_csv(path+file+type_file)
```

## 3.3 Convert to JSON and display the resulting file

Call the convert_df_to_json function, passing the Pandas DataFrame dataframe as an argument. This function converts the DataFrame into a list of JSON-formatted conversations. Display the resulting JSON data.

```python
result_json = convert_df_to_json(dataframe)

print(json.dumps(result_json, indent=2))
```

## 3.4 Save JSON file

Specify path and file type. Stored data into a JSON file.

```python
path = './Cross_Validation_Sample/Cross_Validation_JSON_Files/'
type_file = '.json'

with open(path+file+type_file, 'w') as json_file:
    json.dump(result_json, json_file, indent=2)
```
